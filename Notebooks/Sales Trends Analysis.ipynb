```python
import pandas as pd

df = pd.read_csv('Online_Retail_Cleaned.csv')
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Month_Name'] = df['InvoiceDate'].dt.month_name()
```

```python
monthly_sales = df.groupby(['Year', 'Month', 'Month_Name'])['TotalRevenue'].sum().reset_index()
monthly_sales.head()
```

```python
monthly_sales.sort_values(by='TotalRevenue', ascending=False).head()
```

```python
monthly_sales.sort_values(by='TotalRevenue').head()
```

```python
import matplotlib.pyplot as plt

plt.figure()
monthly_sales.groupby('Month')['TotalRevenue'].sum().plot()
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.title('Monthly Sales Trend')
plt.show()
```

```python
yearly_sales = df.groupby('Year')['TotalRevenue'].sum().reset_index()
yearly_sales
```
